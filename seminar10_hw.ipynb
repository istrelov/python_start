В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?

```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})
data.head(int(data.count()))
```

```python
pd.get_dummies(data)
```

Как я понял задачу, нужно преобразовать DataFeame в такой же вид, как он получается при использовании get_dummies. Для этого я воспользуюсь функцией loc - создам 2 новые колонки с условиями, а потом удалю первую колонку

```python
data.insert(1, 'whoAmI_human', 0)
data.insert(1, 'whoAmI_robot', 0)
data.loc[data['whoAmI']=='human', 'whoAmI_human'] = 1
data.loc[data['whoAmI']=='robot', 'whoAmI_robot'] = 1
del data[data.columns[0]]
data.head(int(data.count()[0]))
```
